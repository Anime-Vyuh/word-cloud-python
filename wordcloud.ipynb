# Word Cloud Python

### Import Libraries

```python
from wordcloud import STOPWORDS,WordCloud
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

### Read the CSV File

```python
file = pd.read_csv("Best 50 anime of all time.csv")
```

```python
data = pd.DataFrame(file)
names = data.Title
```

```python
import random
stopwords = set(STOPWORDS) 

#get the text

name = [anime for anime in data.Title]
    
wordclouds = WordCloud(width = 1200, height = 720, 
                background_color ='white',collocations=False,
                stopwords = stopwords,random_state = 42, 
                max_font_size=60)

wordclouds.generate(''.join(name)) 
  
# plot the WordCloud image 

plt.figure(figsize=(10,8)) 
plt.imshow(wordclouds,interpolation=None) 
plt.axis("off")
plt.tight_layout(pad = 0) 

# save the image 
plt.savefig("anime_word_cloud.png")
```
